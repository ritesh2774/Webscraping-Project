```python
# Scrapping data from makaan.com
import requests
from bs4 import BeautifulSoup

base_url="https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?beds=2&propertyType=villa&page="

l =[]
for page in range(1,22,1):
    #print(base_url+str(page))
    r = requests.get(base_url+str(page))
    c = r.content

    soup = BeautifulSoup(c, "html.parser")

    all = soup.find_all("div",{"class":"cardWrapper"})
    #len(a)
  
    for item in all:
        d={}
        try:
            d["House Type"]=item.find("strong",{"data-type":"listing-link"}).text.replace("\xa0"," ")
        except:
            d["House Type"]=None

        try:
            d["Address"]=item.find("span",{"class":"locName"}).text.replace("\xa0","")
        except:
            d["Address"]=None

        try:    
            d["Price"]=item.find("div",{"data-type":"price-link"}).text
        except:
            d["Price"]=None

        try:
            d["Area(sqrt)"]=item.find("div",{"class":"size"}).text  
        except:
            d["Area(sqrt)"]=None

        try:    
            d["Bahrooms"]=item.find("span",{"title":"Bathrooms"}).text
        except:
            d["Bahrooms"]=None

        try:
            d["Status"]=item.find("div",{"class":"val"}).text
        except:
            d["Status"]=None

        try:
            d["Possession"]=item.find("span",{"title":"Possession by"}).text.replace("\xa0","")
        except:
            d["Possession"]=None

        l.append(d)    
    
```

```python
import pandas
df=pandas.DataFrame(l)
df
```

```python
df.to_csv("Output.csv")
```
